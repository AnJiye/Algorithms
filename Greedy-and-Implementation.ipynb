### â­ <ë¬¸ì œ> 1ì´ ë  ë•Œê¹Œì§€: ë‹µì•ˆ ì˜ˆì‹œ (python)

```python
# N, Kì„ ê³µë°±ì„ ê¸°ì¤€ìœ¼ë¡œ êµ¬ë¶„í•˜ì—¬ ì…ë ¥ ë°›ê¸°
n, k = map(int, input().split())

result = 0

while True:
    # Nì´ Kë¡œ ë‚˜ëˆ„ì–´ ë–¨ì–´ì§€ëŠ” ìˆ˜ê°€ ë  ë•Œê¹Œì§€ ë¹¼ê¸°
    target = (n // k) * k
    result += (n - target)
    n = target
    # Nì´ Kë³´ë‹¤ ì‘ì„ ë•Œ (ë” ì´ìƒ ë‚˜ëˆŒ ìˆ˜ ì—†ì„ ë•Œ) ë°˜ë³µë¬¸ íƒˆì¶œ
    if n < k:
        break
    # Kë¡œ ë‚˜ëˆ„ê¸°
    result += 1
    n //= k

# ë§ˆì§€ë§‰ìœ¼ë¡œ ë‚¨ì€ ìˆ˜ì— ëŒ€í•˜ì—¬ 1ì”© ë¹¼ê¸°
result += (n - 1)
print(result)
```

### â­ <ë¬¸ì œ> 1ì´ ë  ë•Œê¹Œì§€: ë‹µì•ˆ ì˜ˆì‹œ (C++)

```python
#include <bits/stdc++.h>

using namespace std;

int n, k;
int result;

int main(void) {
    cin >> n >> k;
    while (true) {
         # Nì´ Kë¡œ ë‚˜ëˆ„ì–´ ë–¨ì–´ì§€ëŠ” ìˆ˜ê°€ ë  ë•Œê¹Œì§€ ë¹¼ê¸°
        target = (n / k) * k;
        result += (n - target);
        n = target;
        # Nì´ Kë³´ë‹¤ ì‘ì„ ë•Œ (ë” ì´ìƒ ë‚˜ëˆŒ ìˆ˜ ì—†ì„ ë•Œ) ë°˜ë³µë¬¸ íƒˆì¶œ
        if n < k
            break;
        # Kë¡œ ë‚˜ëˆ„ê¸°
        result++;
        n /= k;
    }
    # ë§ˆì§€ë§‰ìœ¼ë¡œ ë‚¨ì€ ìˆ˜ì— ëŒ€í•˜ì—¬ 1ì”© ë¹¼ê¸°
    result += (n - 1);
    cout << result << '\n';
}
```

```python
n = int(input())
data = list(map(str, input().split()))

# L R U D
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]

# ì‹œì‘ ìœ„ì¹˜
x, y = 1, 1

for i in range(n+1):
    if data[i] == 'L':
        nx = x + dx[0]
        ny = y + dy[0]
    elif data[i] == 'R':
        nx = x + dx[1]
        ny = y + dy[1]
    elif data[i] == 'U':
        nx = x + dx[2]
        ny = y + dy[2]
    else:
        nx = x + dx[3]
        ny = y + dy[3]
    
    if 1 <= nx <= n and 1 <= ny <= n:
        x = nx
        y = ny

print(x, y)
```

### ğŸ¤” <ë¬¸ì œ> ìƒí•˜ì¢Œìš°: ë¬¸ì œ í•´ê²° ì•„ì´ë””ì–´

### â­ <ë¬¸ì œ> ìƒí•˜ì¢Œìš°: ë‹µì•ˆ ì˜ˆì‹œ (Python)

```python
# N ì…ë ¥ ë°›ê¸°
n = int(input())
x, y = 1, 1
plans = input().split()

# L, R, U, Dì— ë”°ë¥¸ ì´ë™ ë°©í–¥
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
move_types = ['L', 'R', 'U', 'D']

# ì´ë™ ê³„íšì„ í•˜ë‚˜ì”© í™•ì¸í•˜ê¸°
for plan in plans:
    # ì´ë™ í›„ ì¢Œí‘œ êµ¬í•˜ê¸°
    for i in range(len(move_types)):
        if plan == move_types[i]:
            nx = x + dx[i]
            ny = y + dy[i]
    # ê³µê°„ì„ ë²—ì–´ë‚˜ëŠ” ê²½ìš° ë¬´ì‹œ
    if nx < 1 or ny < 1 or nx > n or ny > n:
        continue
    # ì´ë™ ìˆ˜í–‰
    x, y = nx, ny

print(x, y)
```

### â­ <ë¬¸ì œ> ìƒí•˜ì¢Œìš°: ë‹µì•ˆ ì˜ˆì‹œ (C++)

```python
#include <bits/stdc++.h>

using namespace std;

int n;
string plans;
int x = 1, y = 1;

int dx[4] = {0, 0, -1, 1};
int dy[4] = {-1, 1, 0, 0};
char moveTypes[4] = {'L', 'R', 'U', 'D'};

int main(void) {
    cin >> n;
    cin.ignore(); // ë²„í¼ ë¹„ìš°ê¸° - ì •ìˆ˜ë¥¼ ì…ë ¥ë°›ê³ , ë¬¸ìì—´ì„ ì…ë ¥ë°›ëŠ” ê²½ìš°
    getline(cin, plans);
    
    for (int i = 0 ; i < plans.size() ; i++) {
        char plan = plans[i];
        
        int nx = -1, ny = -1;
        
        for(int j = 0 ; j < 4 ; j++) {
            if (plan == moveTypes[j]) {
                nx = x + dx[j];
                ny = y + dy[j];
            }
        }
        if (nx < 1 || ny < 1 || nx > n || ny > n) continue;
        x = nx;
        y = ny;
    }
    cout << x << ' ' << y << '\n'
    return 0;
}
```
